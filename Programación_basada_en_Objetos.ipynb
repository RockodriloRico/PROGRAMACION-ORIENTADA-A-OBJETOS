```python
# Crear la estructura del Objeto o Clase
# Siempre se inica por la clase de del objeto
'''
 Cada que se realizan cambios en en el codigo de Clase hay que volver ejecutar
 para grabar los cambios
 - SIEMPRE LA CLASE  Lleva la primera letra mayuscula
 '''
class Persona:
#atributos no van a llevar una x
# se le pone una x para covertir un  atributo en una variable
  def __init__ (self, xdni, xnombre, xsexo, xedad) : # la funcion init gurda el valor absoluto de los atributo
    self.dni = xdni
    self.nombre = xnombre
    self.sexo = xsexo
    self.edad = xedad
    return

  def __str__ (self):
     return'La persona: {} \n con el DNI: {} \n de Genero: {} \n tiene {} años de edad: '.format(self.nombre, self.dni, self.sexo, str(self.edad))
  # get recupera el dato
  def getDni (self):
    return self.dni

  def getNombre (self):
    return self.nombre

  def getSexo (self):
    return self.sexo

  def getEdad (self):
    return self.edad
  #set introduce los valores o el dato #Para cambias o remplazar datos de algun objeto ya creado
  def setNombre (selft, xnombre):
    self.nombre = xnombre


```

3.2.1.14 Pila Contadora


```python
class Stack:
    def __init__(self):
        self.__stk = []

    def push(self, val):
        self.__stk.append(val)

    def pop(self):
        val = self.__stk[-1]
        del self.__stk[-1]
        return val


class CountingStack(Stack):
    def __init__(self):
        Stack.__init__(self)
        self.__counter = 0


    def get_counter(self):
        return self.__counter


    def pop(self):
        self.__counter += 1
        return Stack.pop(self)


stk = CountingStack()
for i in range(100):
    stk.push(i)
    stk.pop()
print(stk.get_counter())
```

3.4.1.12 La clase Timer


```python
class Timer:
#Los argumentos del constructor tienen valores predeterminados de 0,
#para el caso en que se omiten los argumentos

    def __init__(self, hours=0, minutes=0, seconds=0):
        self.__hours = hours
        self.__minutes = minutes
        self.__seconds = seconds

#La representación de cadena devuelta por __str__tiene ceros a la izquierda
#agregados cuando es necesario, usando la función format.'''

    def __str__(self):
        return '{:02d}:{:02d}:{:02d}'.format(self.__hours, self.__minutes, self.__seconds)

#El método next_secondaumenta el número de segundos y ajusta los minutos y horas si es necesario,
#asegurándose de que los valores se mantengan dentro de los rangos especificados.

    def next_second(self):
        self.__seconds += 1
        if self.__seconds >= 60:
            self.__seconds = 0
            self.__minutes += 1
            if self.__minutes >= 60:
                self.__minutes = 0
                self.__hours += 1
                if self.__hours >= 24:
                    self.__hours = 0

#El método prev_secondhace lo contrario,
#disminuyendo el número de segundos y ajustando los minutos y horas si es necesario.

    def prev_second(self):
        self.__seconds -= 1
        if self.__seconds < 0:
            self.__seconds = 59
            self.__minutes -= 1
            if self.__minutes < 0:
                self.__minutes = 59
                self.__hours -= 1
                if self.__hours < 0:
                    self.__hours = 23


timer = Timer(23, 59, 59)
print(timer)
timer.next_second()
print(timer)
timer.prev_second()
print(timer)
```

3.4.1.13 Días de la semana


```python
class WeekDayError(Exception):
    pass


class Weeker:
    def __init__(self, day):
        days_of_week = ['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom']
        if day not in days_of_week:
            raise WeekDayError
        self.__day = day

    def __str__(self):
        return self.__day

    def add_days(self, n):
        days_of_week = ['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom']
        current_index = days_of_week.index(self.__day)
        new_index = (current_index + n) % 7
        self.__day = days_of_week[new_index]

    def subtract_days(self, n):
        days_of_week = ['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom']
        current_index = days_of_week.index(self.__day)
        new_index = (current_index - n) % 7
        self.__day = days_of_week[new_index]


try:
    weekday = Weeker('Lun')
    print(weekday)
    weekday.add_days(15)
    print(weekday)
    weekday.subtract_days(23)
    print(weekday)
    weekday = Weeker('Monday')
except WeekDayError:
    print("Lo siento, no puedo atender tu solicitud.")

```

3.4.1.15 Triángulo


```python
import math

class Point:
    def __init__(self, x=0.0, y=0.0):
        self.__x = x
        self.__y = y

    def getx(self):
        return self.__x

    def gety(self):
        return self.__y

    def distance_from_xy(self, x, y):
        return math.hypot(x - self.__x, y - self.__y)

    def distance_from_point(self, point):
        return self.distance_from_xy(point.getx(), point.gety())


class Triangle:
    def __init__(self, vertice1, vertice2, vertice3):
        self.__vertices = [vertice1, vertice2, vertice3]

    def perimeter(self):
        a = self.__vertices[0].distance_from_point(self.__vertices[1])
        b = self.__vertices[1].distance_from_point(self.__vertices[2])
        c = self.__vertices[2].distance_from_point(self.__vertices[0])
        return a + b + c


triangle = Triangle(Point(0, 0), Point(1, 0), Point(0, 1))
print(triangle.perimeter())


```

4.3.1.15 Histograma de frecuencia de caracteres


```python
import string

# Pedir el nombre del archivo de entrada
filename = input("Ingrese el nombre del archivo: ")

# Crear un diccionario para almacenar los recuentos
count_dict = {}

# Abrir el archivo y leerlo línea por línea
with open(filename, "r") as file:
    for line in file:
        # Convertir todas las letras a minúsculas
        line = line.lower()
        # Iterar sobre cada letra en la línea
        for letter in line:
            # Si la letra es una letra latina, aumentar el contador correspondiente
            if letter in string.ascii_lowercase:
                count_dict[letter] = count_dict.get(letter, 0) + 1

# Imprimir el histograma
for letter, count in sorted(count_dict.items()):
    print(f"{letter} -> {count}")

# Crear un archivo de salida con el histograma
with open("histogram.txt", "w") as file:
    for letter, count in sorted(count_dict.items()):
        file.write(f"{letter} -> {count}\n")
```

4.3.1.16 Histograma de frecuencia de caracteres ordenado


```python
import os

filename = input("Ingrese el nombre del archivo: ")
if not os.path.isfile(filename):
    print("El archivo no existe.")
else:
    with open(filename, 'r') as f:
        text = f.read().lower()

    frequencies = {}
    for char in text:
        if char.isalpha():
            if char in frequencies:
                frequencies[char] += 1
            else:
                frequencies[char] = 1

    sorted_freq = sorted(frequencies.items(), key=lambda x: x[1], reverse=True)

    hist_filename = os.path.splitext(filename)[0] + '.hist'
    with open(hist_filename, 'w') as f:
        for char, freq in sorted_freq:
            f.write(f"{char} -> {freq}\n")
            print(f"{char} -> {freq}")

    print(f"Histograma guardado en {hist_filename}")
```

4.3.1.17 Evaluando los resultados de los estudiantes


```python
class EmptyFileError(Exception):
    pass

class InvalidLineError(Exception):
    pass

def read_file(filename):
    students = {}
    try:
        with open(filename, "r") as file:
            lines = file.readlines()
            if not lines:
                raise EmptyFileError("El archivo está vacío")
            for line in lines:
                data = line.strip().split()
                if len(data) != 3:
                    raise InvalidLineError(f"Línea incorrecta: {line}")
                name, surname, points = data
                key = (name, surname)
                if key not in students:
                    students[key] = 0
                students[key] += float(points)
    except FileNotFoundError:
        print(f"Archivo no encontrado: {filename}")
    except EmptyFileError as e:
        print(e)
    except InvalidLineError as e:
        print(e)
    else:
        return students

def print_report(students):
    for student, total_points in sorted(students.items(), key=lambda x: (-x[1], x[0])):
        print(f"{student[0]} {student[1]} \t {total_points}")


filename = input("Ingresa el nombre del archivo: ")
students = read_file(filename)
if students:
    print_report(students)
```

```python
from calendar import Calendar


class MyCalendar(Calendar):

    def count_weekday_in_year(self, year, weekday):
        count = 0
        for month in self.monthdays2calendar(year, 1):
            for week in month:
                if week[weekday] != 0:
                    count += 1
        return count


# Ejemplos de uso
cal = MyCalendar()

# Ejemplo 1
year = 2019
weekday = 0
print(f"El número de Lunes en el año {year} es {cal.count_weekday_in_year(year, weekday)}")



```
